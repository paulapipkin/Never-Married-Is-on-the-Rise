Import Libaries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector,  make_column_transformer
#from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

Loading data

```python
df = pd.read_csv('/content/drive/MyDrive/BootCamp/05 - 08 - MACHINE LEARNING/01 Week 05 - Intro to Machine Learning/abalone.data', header=None)
df.head()
```

adding columns names

```python
df.columns=['Sex','Length','Diameter','Height','Whole weight','Shucked weight',	'Viscera weight',	'Shell weight','Rings']
df.head()
```

```python
df.duplicated().sum()
```

1) Separate your data into the features matrix (X) and target vector (y).

2) Train/test split the data. Please use the random number 42 for consistency.




```python
X = df.drop(columns='Rings')
y = df['Rings']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

3) Create a ColumnTransformer to preprocess the data. Remember to:

    a) Create column selectors for the numeric and categorical columns

    b) Create a OneHotEncoder for one-hot encoding the categorical columns

    c) Create a StandardScaler for scaling numeric columns

    d) Match each transformer with the appropriate selector in a tuple

    e) Use the tuples to create a ColumnTransformer to preprocess the data.



```python
#a) Instantiate the categorical and numeric column selectors
cat_selector = make_column_selector(dtype_include='object')
num_selector = make_column_selector(dtype_include='number')
```

```python
# b & c) Instantiate the Standard Scaler and the One hot encoder
ohe_encoder = OneHotEncoder(handle_unknown='ignore')
scaler = StandardScaler()
```

```python
#d) Make tuples for preprocessing the categorical and numeric columns
num_col_tuple = (scaler, num_selector)
cat_col_tuple = (ohe_encoder,cat_selector)
```

```python
#e)Use the tuples to create a ColumnTransformer to preprocess the data
transformer = make_column_transformer(num_col_tuple, cat_col_tuple, remainder='passthrough')
```

4) Transform your data and display the result.


```python
transformer.fit(X_train)
```

```python
X_train_processed = transformer.transform(X_train)
X_test_processed = transformer.transform(X_test)
```

```python
X_train_df = pd.DataFrame(X_train_processed)
X_train_df.head()
```
