##Data Visualization Exercise

```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename= 'https://query.data.world/s/m65tedw6z6qe3dvvlaag2xmafqqxsg'
df= pd.read_excel(filename)
df.head()
```

```python
df.sort_values(by=['Year','Age Range'], inplace=True, ignore_index=True)
df.head()
```

```python
males = df.loc[df['Gender']=='Men']
females = df.loc[df['Gender']=='Women']
```

```python
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(14,5))
fig.subplots_adjust(wspace=0.5)

#males
sns.lineplot(data= males, x='Year', y='Never Married', hue='Age Range', ax=axes[0])
axes[0].set_xticks(np.arange(2006, 2017, 10.0))
axes[0].legend(title='Age Range' , loc=(1.02,0))
axes[0].grid(axis='y', c='gray', alpha = .5, linestyle = '--')
axes[0].set_xlabel('2006 vs 2016', fontsize = 16)
axes[0].set_ylabel('Percentage of "Never Married"', fontsize = 16)
axes[0].set_title('Change on Never Married Rate for Males', fontsize = 16);
#females
sns.lineplot(data= females, x='Year', y='Never Married', hue='Age Range', ax=axes[1])
axes[1].set_xticks(np.arange(2006, 2017, 10.0))
axes[1].legend(title='Age Range' ,loc=(1.02,0))
axes[1].grid(axis='y', c='gray', alpha = .5, linestyle = '--')
axes[1].set_xlabel('2006 vs 2016', fontsize = 16)
axes[1].set_ylabel('Percentage of "Never Married"', fontsize = 16)
axes[1].set_title('Change on Never Married Rate for Females', fontsize = 16);
```

Bar Plot

```python
g = sns.catplot(data=df, x='Age Range', y='Never Married', hue='Year', col='Gender', kind='bar',height=4, aspect=.7) 
# from https://seaborn.pydata.org/generated/seaborn.barplot.html
```

```python
group1 = df.loc[df['Age Range']=='25-29']
group2 = df.loc[df['Age Range']=='30-34']
group3 = df.loc[df['Age Range']=='35-44']
group4 = df.loc[df['Age Range']=='45-54']
group5 = df.loc[df['Age Range']=='55+']
```

```python
fig1, axes1 = plt.subplots(3, 2, figsize=(18, 9))
fig.subplots_adjust(wspace=0.5, hspace=0.7, bottom=0.4, top= 0.8)

sns.barplot(data=group1, x='Gender', y='Never Married', hue='Year', ax=axes1[0,0]).set_title('Age Range 25-29')
axes1[0,0].set_xlabel('')
sns.barplot(data=group2, x='Gender', y='Never Married', hue='Year', ax=axes1[0,1]).set_title('Age Range 30-34')
axes1[0,1].set_xlabel('')
sns.barplot(data=group3, x='Gender', y='Never Married', hue='Year', ax=axes1[1,0]).set_title('Age Range 35-44')
axes1[1,0].set_xlabel('')
sns.barplot(data=group4, x='Gender', y='Never Married', hue='Year', ax=axes1[1,1]).set_title('Age Range 45-54')
axes1[1,1].set_xlabel('')
sns.barplot(data=group5, x='Gender', y='Never Married', hue='Year', ax=axes1[2,0]).set_title('Age Range 55+')
axes1[2,0].set_xlabel('');
```
